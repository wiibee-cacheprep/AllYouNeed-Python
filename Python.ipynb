#### Programming Language
This captures the essence of programming for me - “we do things with stuff”!

“Things” take the form of operations like addition and concatenation, and “stuff” refers to the objects on which we perform those operations.


Everything is an object in a Python script. Objects are like alphabets for Python.

More formally, objects are essentially just pieces of memory, with values and sets of associated operations.

e.g. a = 90 creates an object ‘a’ as type Integer, value ‘90’ and supported operations of  addition, subtraction etc.

Hierarchy of Python:
Python programs can be decomposed into modules, statements, expressions, and objects, as follows:
1. Programs are composed of modules. 
2. Modules contain statements.
3. Statements contain expressions.
4. Expressions create and process objects.
